````python
import pandas as pd
names_series = pd.read_excel('/home/tha/ghub/4sem/python/2020s/unisex_navne.xls',header=None,names=['navne'])['navne']
names_list = list(names_series)
def get_names(ls):
    current = 0
    while(current < len(ls)):
        yield ls[current]
        current += 1

for name in get_names(names_list):
    print(name)
````

````python
import json
import api_keys
import requests
import pandas as pd

url = "https://api.openweathermap.org/data/2.5/forecast"
query = {'q': 'Copenhagen,dk', 
         'mode': 'json',                       
         'units': 'metric',
         'appid': api_keys.OWM_API_KEY}
r = requests.get(url, params=query)

data = r.json()
data['list']
````

````python
def get_main_data(url, query):
    result = []
    r = requests.get(url, params=query)
    data = r.json()
    ls = data['list']
    for item in ls:
        main = item['main']
        result.append(main)
    return result
def get_min_max_temp(city_code):
    query = {'q': city_code, 
         'mode': 'json',                       
         'units': 'metric',
         'appid': api_keys.OWM_API_KEY}
    all_mains = get_main_data(url, query)
    df = pd.DataFrame(all_mains)
    min_val = min(df['temp_min'])
    max_val = max(df['temp_max'])
    return min_val,max_val
min_val, max_val = get_min_max_temp('Copenhagen,DK')
print('min temp:',min_val,'\nmax temp:',max_val)
````

````python
from collections import defaultdict

with open('cities.json','r') as jsonfile:
    cities=json.load(jsonfile)
    print('{} cities \nThe first 10 are: {}'.format(len(cities),cities[:10]))
    codes = defaultdict(list)
    for city in cities:
        #codes.append((city['name'],city['country']))
        codes[city['country']].append('{},{}'.format(city['name'],city['country']))
print('')
print('{} byer i Danmark'.format(len(codes['DK'])))
print(codes['DK'])
````

````python
# create a function to get average temp from all cities in a country
def get_avg_temp(citylist):
    temps = []
    for city in citylist:
        avg_temp = sum(get_min_max_temp(city))/2
        temps.append(avg_temp)
    return sum(temps)/len(temps)
get_avg_temp(codes['DK'])
````

* Write a function that creates a URL by adding an index to this url:
    `https://api.github.com/repositories/33015583/contributors?page=`
* Create a list of 100 URLs by calling the above function 100 times with indices from 0 to 99 inclusive
* Using `%%timeit`, measure how long it takes to call the url 100 times sequentially
* Create a thread pool
* Using `%%timeit`, measure how long it takes to call the url 100 times *in parallel*

````python
import webget
url = 'https://api.github.com/repositories/33015583/contributors?page='
urls = [url + str(idx) for idx in range(500)]
````

````python
def download_sequential():
    for url in urls:
        webget.download(url,'todelete.txt')
````

````python
%%timeit
download_sequential()
````

````python
from multiprocessing import Pool
from functools import partial
from concurrent.futures import ThreadPoolExecutor
````

````python
%%timeit

def download_paralel():
    with Pool(4) as pool:
        download_todelete = partial(webget.download,to='todelete.txt')
        pool.map(download_todelete,urls)

download_paralel()
````

````python
%%timeit
def download_concurrent():
    with ThreadPoolExecutor(max_workers=10) as executor:
        for url in urls:
            future = executor.submit(partial(webget.download,to='todelete.txt'), url) 

download_concurrent()
````

# Exercise with threads and processes
1. Create a function that can take a list of image urls with http and using a thread pool can download the images to disk
  - test the function with dummy images from https://dummyimage.com/600x400/000/fff (/image_size/background_color/foreground_color can be set programmatically to load many different images)
2. Create a function that can use process pool to calculate fibonacci sequence using this recursive fibonacci function:
```python
def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))
```
  - Make your function so that it can take a sequence of numbers and return the corresponding fibbonaci values   
  
HINT: the processor time needed for each calculation increases exponentially as higher numbers are given to the function. Therefore it makes sense to create processes equivalent to amount of cpu cores on the computer to distribute the work load as it gets really heavy moving beyond 35 (on my computer)

````python
def get_img_urls():
    urls, filenames = [],[]
    for height in range(100,600,200):
        for width in range(100,600,200):
            for background in range(0,16,2):
                for foreground in range(0,16,2):
                    # print('{:X}'.format(i)*3)
                    urls.append('https://dummyimage.com/{width}x{height}/{background:x}{background:x}{background:x}/{foreground:x}{foreground:x}{foreground:x}'
                                .format(height=height,width=width,background=background,foreground=foreground))
                    filenames.append('img{width}x{height}-{background:x}{background:x}{background:x}-{foreground:x}{foreground:x}{foreground:x}.png'
                                .format(height=height,width=width,background=background,foreground=foreground))
    return urls,filenames
````

````python
# 1 Thread pool exercise
import requests, os 
from concurrent.futures import ThreadPoolExecutor
urls, filenames = get_img_urls()
def save_img_from_url(url, filename):
    """given a image url requests the image and store it on filename"""
    ## add something like: 
    filename = f'todelete/imgs/{filename}'
    if not os.path.exists(os.path.dirname(filename)):
        try:
            os.makedirs(os.path.dirname(filename))
        except OSError as exc: # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise
    with open(filename,'wb') as f: 
        f.write(requests.get(url).content)
    return f'{filename} was created' 

with ThreadPoolExecutor() as pool:
    results = pool.map(save_img_from_url,urls,filenames) 
    for e in results:
        print(e)
````

````python
def fibo(n):
    if n <= 1:
        return n
    else:
        return(fibo(n-1) + fibo(n-2))
````

````python
# 2 process pool exercise
from concurrent.futures import ProcessPoolExecutor
def calc_fibo(seq):
    with ProcessPoolExecutor() as executor:
        results = list(executor.map(fibo, seq))
        return results
````

````python
%%timeit
calc_fibo([2,3,4,20,34,35])
````
